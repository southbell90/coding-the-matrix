```python
strlist = [x.split() for x in mylist]
```

### Task 2.12.1

```python
def create_voting_dict(strlist):
    map_name_voting = {}
    for s in range(len(strlist)):
        voting_record = []
        for i in range(len(strlist[s])):
            if i >= 0 and i <= 2:
                continue
            voting_record.append(int(strlist[s][i]))
        map_name_voting[strlist[s][0]] = voting_record

    return map_name_voting
```

### Task 2.12.2

```python
def policy_compare(sen_a, sen_b, voting_dict):
    simty = 0
    for i in range(len(voting_dict[sen_a])):
        simty += (voting_dict[sen_a][i] * voting_dict[sen_b][i])

    return simty
```

### Task 2.12.3

```python
def most_similar(sen, voting_dict):
    most_simty = -10e8
    most_name = ''
    for name, voting in voting_dict.items():
        if name == sen:
            continue
        now_simty = policy_compare(sen, name, voting_dict);
        if now_simty > most_simty:
            most_simty = now_simty
            most_name = name
    return most_name
        
```

### Task 2.12.4

```python
def least_similar(sen, voting_dict):
    least_simty = 10e8
    least_name = ''
    for name, voting in voting_dict.items():
        if name == sen:
            continue
        now_simty = policy_compare(sen, name, voting_dict);
        if now_simty < least_simty:
            least_simty = now_simty
            least_name = name
    return (least_name, least_simty)
```

### Task 2.12.5

```python
voting_dict = create_voting_dict(strlist)
most_Rhode = most_similar('Chafee', voting_dict)
least_Rick = least_similar('Santorum', voting_dict)
print(most_Rhode)
print(least_Rick)
```

### Task 2.12.7

```python
def find_average_similarity(sen, sen_set, voting_dict):
    accum = 0
    for sen_name in sen_set:
        accum += policy_compare(sen, sen_name[0], voting_dict)
    accum /= len(sen_set)

    return accum

democrat_list = [demo for demo in strlist if demo[1] == 'D']


greatest_avg_sim = float("-inf")
greatest_name = ""
for sen in strlist:
    now_avg_sim = find_average_similarity(sen[0], democrat_list, voting_dict)
    if now_avg_sim > greatest_avg_sim:
        greatest_avg_sim = now_avg_sim
        greatest_name = sen[0]

print(greatest_name, greatest_avg_sim)
```

### Task 2.12.8

```python
def find_average_record(sen_set, voting_dict):
    result_list = [0] * len(list(voting_dict.values())[0])
    for sen in sen_set:
        result_list = [x + y for x,y in zip(result_list, voting_dict[sen[0]])]
    size = len(sen_set)
    return [x / size for x in result_list]

average_Democrat_record = find_average_record(democrat_list, voting_dict)

most_avg_name = ""
most_simty = float("-inf")
for sen in strlist:
    now_simty = 0
    for i in range(len(voting_dict[sen[0]])):
        now_simty += (voting_dict[sen[0]][i] * average_Democrat_record[i])
    if now_simty > most_simty:
        most_simty = now_simty
        most_avg_name = sen[0]

print(most_avg_name, most_simty)
```

### Task 2.12.9

```python
def bitter_rivals(voting_dict):
    sen_a = ""
    sen_b = ""
    least_sim = float("inf")
    for key in voting_dict:
        name_sim = least_similar(key, voting_dict)
        if name_sim[1] < least_sim:
            sen_a = name_sim[0]
            sen_b = key
            least_sim = name_sim[1]
    return (sen_a,sen_b)


bitter_rivals(voting_dict)        
```
