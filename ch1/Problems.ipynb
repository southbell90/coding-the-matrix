## Python comprehension problems

### Problem 1.7.1

```python
def my_filter(L, num): return [x for x in L if x % num != 0]
```

```python
my_filter([1,2,4,5,7], 2)
```

### Problem 1.7.2

```python
def my_lists(L): return [[x for x in range(y+1) if x != 0 ] for y in L]
```

```python
my_lists([1,2,4])
```

```python
my_lists([0])
```

### Problem 1.7.3

```python
def my_function_composition(f,g): return {key: g[value] for key, value in f.items() }
```

```python
f =  {0:'a', 1:'b'}
g = {'a':'apple', 'b':'banana'}
my_function_composition(f,g)
```

### Problem 1.7.4

```python
def mySum(L):
    current = 0
    for x in L:
        current += x
    return current
```

```python
mySum([1,2,3,4,5])
```

### Problem 1.7.5

```python
def myProduct(L):
    current = 1
    for x in L:
        current *= x
    return current
```

```python
myProduct([1,2,4])
```

### Problem 1.7.6

```python
def myMin(L):
    current = float('inf')
    for x in L:
        if x < current:
            current = x
    return current
```

```python
myMin([1,2,3])
```

### Problem 1.7.7

```python
def myConcat(L):
    str = ''
    for x in L:
        str += x
    return str
```

```python
myConcat(['aa','b ','c'])
```

### Problem 1.7.8

```python
def myUnion(L):
    current = set()
    for x in L:
        current = current.union(x)
    return current
```

```python
myUnion([{1,2}, {2,3}])
```
