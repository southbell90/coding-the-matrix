### Task 4.15.2

```python
def identity():
    return Mat(({'x','y','u'},{'x','y','u'}), {('x','x'): 1, ('x','y'):0, ('x','u'):0, ('y','x'):0, ('y','y'):1, ('y','u'):0, ('u','x'): 0, ('u','y'):0, ('u','u'):1})
```

### Task 4.15.3

```python
def translation(alpha, beta):
    return Mat(({'x','y','u'},{'x','y','u'}), {('x','x'): 1, ('x','y'):0, ('x','u'):alpha, ('y','x'):0, ('y','y'):1, ('y','u'):alpha, ('u','x'): 0, ('u','y'):0, ('u','u'):1})
```

### Task 4.15.4

```python
def scale(alpha, beta):
    return Mat(({'x','y','u'},{'x','y','u'}), {('x','x'): alpha, ('x','y'):0, ('x','u'):0, ('y','x'):0, ('y','y'):beta, ('y','u'):0, ('u','x'): 0, ('u','y'):0, ('u','u'):1})
```

### Task 4.15.5

```python
def rotation(theta):
    Mat(({'x','y','u'},{'x','y','u'}), {('x','x'): math.cos(theta), ('x','y'):-math.sin(theta), ('x','u'):0, ('y','x'):math.sin(theta), ('y','y'):math.cos(theta), ('y','u'):0, ('u','x'): 0, ('u','y'):0, ('u','u'):1})
```

### Task 4.15.6

```python
def rotation_about(theta, x, y):
    return translation(x,y)*rotation(theta)*translation(-x,-y)
```

### Task 4.15.7

```python
def reflect_y():
    return Mat(({'x','y','u'},{'x','y','u'}), {('x','x'): -1, ('x','y'):0, ('x','u'):0, ('y','x'):0, ('y','y'):1, ('y','u'):0, ('u','x'): 0, ('u','y'):0, ('u','u'):1})
```

### Task 4.15.8

```python
def reflect_x():
    return Mat(({'x','y','u'},{'x','y','u'}), {('x','x'): 1, ('x','y'):0, ('x','u'):0, ('y','x'):0, ('y','y'):-1, ('y','u'):0, ('u','x'): 0, ('u','y'):0, ('u','u'):1})
```

### Task 4.15.9

```python
def scale_color(r, g, b):
    return Mat(({'r','g','b'},{'r','g','b'}), {('r','r'): r, ('r','g'):0, ('r','b'):0, ('g','r'):0, ('g','g'):g, ('g','b'):0, ('b','r'): 0, ('b','g'):0, ('b','b'):b})
```

### Task 4.15.10

```python
def grayscale():
    return Mat(({'r','g','b'},{'r','g','b'}), {('r','r'): 77/256, ('r','g'):151/256, ('r','b'):28/256, ('g','r'):77/256, ('g','g'):151/256, ('g','b'):28/256, ('b','r'): 77/256, ('b','g'):151/256, ('b','b'):28/256})
```

### Task 4.15.11
