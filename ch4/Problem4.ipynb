### Problem 4.17.13

```python
def lin_comb_mat_vec_mult(M,v):
    accum_vec = Vec(M.D[0], {})
    for i in M.D[1]:
        v_ = Vec(M.D[0])
        for j in M.D[0]:
            v_[i] = v[i] * M[j,i]
        accum_vec = accum_vec + v_
    return accum_vec
        
```

### Problem 4.17.14

```python
def lin_comb_vec_mat_mult(v,M):
    accum_vec = Vec(M.D[1], {})
    for i in M.D[0]:
        v_ = Vec(M.D[1])
        for j in M.D[1]:
            v_[i] = v[i] * M[i,j]
        accum_vec = accum_vec + v_

    return accum_vec
```

### Problem 4.17.15

```python
def dot_product_mat_vec_mult(M,v):
    accum_vec = {M.D[0], {}}
    mat_row = matutil.mat2rowdict(M)
    for i in M.D[0]:
        accum_vec = accum_vec + v*mat_row[i]
    return accum_vec
    
```

### Problem 4.17.16

```python
def dot_product_vec_mat_mult(v, M):
    accum_vec = {M.D[1], {}}
    mat_col = matutil.mat2coldict(M)
    for i in M.D[1]:
        accum_vec = accum_vec + v*mat_col[i]
    return accum_vec
```

### Problem 4.17.17

```python
def Mv_mat_mat_mult(A,B):
    col_dict = {}
    col_vec = matutil.mat2coldict(B)
    for i in B.D[1]:
        col_dict[i] = A * col_vec[i]

    return matutil.coldict2mat(col_dict)
    
```

### Problem 4.17.18

```python
def vM_mat_mat_mult(A,B):
    row_dict = {}
    row_vec = matutil.mat2rowdict(A)
    for i in A.D[0]:
        row_dict[i] = row_vec[i] * B

    return matutil.rowdict2mat(row_dict)
```

### Problem 4.17.19

```python
f = open('UN_voting_data.txt')
```

```python
un_list = list(f)
```

```python
strlist = [x.split() for x in un_list]
```

```python
col_dict = {}
domain_size = len(strlist[0]) - 1
for l in strlist:
    country = ''
    col_vec = Vec({a for a in range(domain_size)},{})
    for i, s in enumerate(l):
        if i == 0:
            country = s
            continue
        col_vec[i - 1] = int(s)
    col_dict[country] = col_vec

un_mat = matutil.coldict2mat(col_dict)
```

```python
M = un_mat.transpose() * un_mat
```

```python
sorted([(value, key) for key,value in M.f.items()])
```
