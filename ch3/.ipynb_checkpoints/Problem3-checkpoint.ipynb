### Problem 3.8.1

```python
def vec_select(veclist, k):
    return [v for v in veclist if v[k] == 0]
```

```python
vec_select([Vec({0,1}, {0:0,1:1}), Vec({0,1}, {0:0})], 1)
```

```python
def vec_sum(veclist, D):
    dic = {}
    for domain in D:
        dic[domain] = 0
    return sum(veclist, Vec(D, dic))
```

```python
vec_sum([], {'a','b','c'})
```

```python
vec_sum([Vec({'a','b'}, {'a':1,'b':2}), Vec({'a','b'},{'a':3, 'b':4})], {'a','b'})
```

```python
def vec_select_sum(D, veclist, k):
    v_k = vec_select(veclist, k)
    return vec_sum(v_k, D)
```

```python
vec_select_sum({'a','b'},[Vec({'a','b'}, {'a':0,'b':2}), Vec({'a','b'},{'a':3, 'b':4})],'a')
```

### Problem 3.8.2

```python
def scale_vec(vecdict):
    return [ v/k for k, v in vecdict.items()]
```

```python
scale_vec({1:Vec({'a','b'}, {'a':1,'b':2}) , 2:Vec({'a','b'},{'a':3, 'b':4})})
```

### Problem 3.8.3

```python
from itertools import combinations

def GF2_span(D, L):
    ret = [Vec(D, {})]
    for i in range(len(L)+1):
        if i == 0:
            continue
        iter_list = [list(comb) for comb in combinations(L, i)]
        for il in iter_list:
            ret.append(il)
    return ret

    
    
    
```

```python
GF2_span({'a','b'}, [Vec({'a','b'}, {'a':0,'b':0}), Vec({'a','b'}, {'a':1,'b':1}), Vec({'a','b'}, {'a':0,'b':1})])
```
